```python
# Library import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Inferential Statistics

In this section, we will approach the basic concepts of populations and sample data, and explore how distributions and their characteristics (such as the median and standard error) can be used to draw inferential conclusions.

For this section, we will simulate normal population data. To do this, we will create random data as follows:

```python
# Set seed for reproducibility
np.random.seed(1234)

# Create a normal population data (median = 100, standard deviation = 2, number of data points = 100.000)
population = np.random.normal(loc=100, scale=2, size=100000)

# plot population data
plt.hist(population, bins=50, edgecolor='black')
plt.title("Population Histogram")
plt.xlabel("Value")
plt.ylabel("Frecuency")
plt.show()
```
